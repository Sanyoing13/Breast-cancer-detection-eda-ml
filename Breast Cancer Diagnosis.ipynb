## **Breast Cancer Diagnosis: EDA and Predictive Modeling**

This project analyses a breast cancer dataset using Exploratory Data Analysis (EDA) and Machine Learning to build a model that predicts whether a tumor is malignant or benign based on features derived from digitized image of a breast mass.

**Objective**

To identify important patterns in the data and build a reliable classification model to assist in early cancer detection.

**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

**Load and Convert the Dataset**

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df['target'] = data.target
```

**Distribution of Selected Features**

```python
features = ['mean radius', 'mean texture', 'mean area', 'mean smoothness']

for feature in features:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=feature, hue='target', kde=True)
    plt.title(f'{feature} Distribution by Target')
    plt.show()
```

**Boxplots to Detect Outliers**

```python
plt.figure(figsize=(10,6))
sns.boxenplot(data=df[features])
plt.title('Boxplot of Selected Features')
plt.xticks(rotation=45)
plt.show()
```

**Pairplot**

```python
sns.pairplot(df[['mean radius', 'mean texture', 'mean area', 'mean smoothness', 'target']], hue='target')
plt.show()
```

### **Feature Scaling:**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Train_Test_Split:**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
np.std(X_train_scaled).round()
```

```python
x1 = 15.46
x_mean = 14.18
std_x = 3.57

(x1-x_mean)/std_x
```

**Model Building**

Train and evaluate multiple models using standard classification metrics.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

**Model Evaluation**

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, roc_auc_score, roc_curve, classification_report
```

```python
sns.countplot(x = y)
plt.show()
```

```python
acc_sc = accuracy_score(y_test, y_pred)
rec_sc = recall_score(y_test, y_pred)
prec_sc = precision_score(y_test, y_pred)
cr = classification_report(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
```

```python
print(f"accuracy_score: {acc_sc}\nrecall_sore: {rec_sc}\nprecision score:\n{prec_sc}\nclassification_report:\n{cr}\nconfusion_matrix:\n{cm}")
```

**The model is a good fit since all scores are around 99%.**

```python
ras = roc_auc_score(y_test, y_pred)
```

```python
ras
```

**A area-under-curve (AUC) score oof 0.99 indicates a good model.**

roc_auc_score measures overall model discrimination ability. **Value around 0.5 means NO Discrimination. Value > 0.7 is considered as good.**

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
```

```python
plt.figure(figsize=(4,2))
plt.plot(fpr, tpr)
plt.plot([0,1], [0,1])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```

**Since the roc curve is above the diagonal and closer to y-axis, it implies higher tpr for the given thresholds, hence is a good fit.**

```python
coef = pd.Series(model.coef_[0], index = X.columns)
np.exp(coef)
```

- A value of np.exp(coef) < 1 implies that the probability of positive class reduces by 1-np.exp(coef).
- On the other hand a value > 1 implies that the pprobability of positive class increases by np.exp(coef)-1.
